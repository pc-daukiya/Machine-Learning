Bar Plot Python Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('penguins')
df
```

```python
sns.barplot(x=df.island, y=df.bill_length_mm)
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue = "sex")
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df, hue = "sex", order= order_1, hue_order=["Female","Male"], ci=90, n_boot= 2)
# ci interval ki size ko badane ke liye ya kam karne ke liye use hota hai
# combinational interval ko yadi gap ke sath Ya accurate dikhana hai toh n_boot=2 ya other value change karke dikha sakte hai
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=df, hue ="sex", orient="v" )
# orient ka use plot ko vertical se horizental me karne ke liye hota hai 
# or esme x or y both axies me numeric values honi chahiye string elements not allowed
plt.show()
```

```python
# sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=df, orient="h" )
# sns.barplot(x="island", y="bill_length_mm", hue="sex", data=df, color="g", saturation=10 )
sns.barplot(x="island", y="bill_length_mm", hue="sex", data=df, saturation=10, errcolor="b", dodge=True)
# dodge ko bars ko overlap ke liye use kiya jata hai

plt.show()
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island", y="bill_length_mm", hue="sex", data=df, saturation=10, errcolor="b", alpha = 0.9)
plt.show()
```
