```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7,8]
var_1 = [3,4,5,1,2,3,6,9]

x_1 = pd.DataFrame({"var": var, "var_1": var_1})


sns.lineplot(x="var", y="var_1",data=x_1)
```

```python
plt.plot(var, var_1)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()


sns.set_theme(style="whitegrid")

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker")
plt.title("Total Bill vs. Tip (Colored by Smoker)")
plt.show()
```

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line Plot of Total Bill by Size")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, ci="sd", palette="Blues")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.histplot(tips["total_bill"], kde=True, color="skyblue")
plt.title("Distribution of Total Bill")
plt.show()
```
