```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset("penguins")
data
```

```python
sns.histplot(data["flipper_length_mm"], color="skyblue")
plt.title("Histogram Plot")
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], color="r", bins = [170,180, 190, 200, 210, 220, 230, 240], kde = True, rug = True)
# kernal density estimation [kde], ka use plots ke uppur wave banane ke liye hota hai kde = True ka use kar sakte hai.
# kde ka scale me show karne ke liye rug = True ka use kar sakte hai.
plt.title("Histogram Plot")
plt.show()
```

```python
sns.displot(data["flipper_length_mm"], color="r", log_scale= True)
plt.title("Histogram Plot")
plt.show()
```
