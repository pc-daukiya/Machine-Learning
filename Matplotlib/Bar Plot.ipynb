```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85,70, 62, 82 ]
z= [20,34,45,67]

width=0.2
p=np.arange(len(x))   
p1=[j+width for j in p]

plt.xlabel("language", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Bar Plot", fontsize=15)

plt.bar(p,y,width,color="r", label="popularity")
plt.bar(p1,z,width,color="y", label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85,70, 62, 82 ]
z= [20,34,45,67]

plt.ylabel("language", fontsize=15)
plt.xlabel("No.", fontsize=15)
plt.title("Bar Plot", fontsize=15)

plt.barh(x,y,width,color="r", label="popularity")
plt.barh(x,z,width,color="y", label="popularity1")

plt.legend()
plt.show()
```
